```python
import numpy as np
# [[Row 1], [Row 2]]
matrix = np.array([[1, 2,3],
                   [3, 4,5]])

# Full array ko print karte hain
print("Yeha hai humara 2D Array:")
print(matrix)

# Ek single value nikalne ke liye

print(matrix[1, 2])
```
