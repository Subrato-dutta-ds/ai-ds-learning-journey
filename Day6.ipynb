```python
class BankAccount:
    def __init__(self, initial_balance):
        # Initialize with an initial balance
        self.__balance = initial_balance
        # Optional: Keep track of total deposits and withdrawals
        self.total_deposits = 0
        self.total_withdrawals = 0

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            self.total_deposits += amount
            print(f"Deposited {amount}. New balance: {self.__balance}")
        else:
            print("Deposit amount must be positive.")

    def withdraw(self, amount):
        if amount <= 0:
            print("Withdrawal amount must be positive.")
        elif amount <= self.__balance:
            self.__balance -= amount
            self.total_withdrawals += amount
            print(f"Withdrew {amount}. New balance: {self.__balance}")
        else:
            print("Insufficient balance")

    def get_balance(self):
        return self.__balance

# Create a BankAccount object with an initial balance of 1000
b = BankAccount(1000)

# Perform a deposit
b.deposit(2000)

# Perform a withdrawal
b.withdraw(1200)

# Check final balance
print(f"Final balance: {b.get_balance()}")
```

```python
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance   # Private variable

    def deposit(self, amount):
        self.__balance += amount
        print(f"₹{amount} deposited")

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
            print(f"₹{amount} withdrawn")
        else:
            print("Insufficient Balance")

    def get_balance(self):
        return self.__balance


# Object Creation
acc = BankAccount(1000)

acc.deposit(500)
acc.withdraw(200)

print("Current Balance:", acc.get_balance())
```

```python
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        # __balance is a PRIVATE variable (Cannot be accessed directly outside the class)
        self.__balance = balance

    # Method to check the balance safely (Getter)
    def show_balance(self):
        return f"Account Owner: {self.owner} | Current Balance: ${self.__balance}"

    # Method to update the balance safely (Setter)
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"${amount} deposited successfully!")
        else:
            print("Error: Invalid deposit amount!")

# Creating an Object
account1 = BankAccount("Abhishek", 5000)
account2 = BankAccount("Rahul", 7000)

# 1. Trying to access the private variable directly (This will throw an AttributeError)
#print(account.__balance)

# 2. Correct way to check the balance using the method
print(account1.show_balance())
print(account2.show_balance())

# 3. Depositing money using the setter method
account1.deposit(2000)

# 4. Checking the updated balance
print("\nAfter Deposit:")
print(account1.show_balance())
print(account2.show_balance())
```
