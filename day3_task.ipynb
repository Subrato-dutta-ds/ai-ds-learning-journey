```python
import json

# JSON File Load
with open('/content/data.json', 'r', encoding='utf-8') as file:
    data = json.load(file)

print(f"\nTotal Products Found: {len(data)}\n")

for index, item in enumerate(data, start=1):

    # Basic Information
    name = item.get("name", "N/A")
    category = item.get("sub_sub_category_name", "N/A")
    price = item.get("min_catalog_price", "N/A")
    product_price = item.get("min_product_price", "N/A")
    description = item.get("description", "N/A")
    hero_product_name = item.get("hero_product_name", "N/A")

    # Rating Information
    reviews_data = item.get("catalog_reviews_summary", {})
    rating = reviews_data.get("average_rating", "N/A")
    review_count = reviews_data.get("review_count", "N/A")
    rating_count = reviews_data.get("rating_count", "N/A")

    # Product Details
    full_details = item.get("full_details", "N/A")

    # Image URL
    image_url = item.get("image", "N/A")

    print("=" * 120)
    print(f"PRODUCT #{index}")
    print("=" * 120)

    print(f"Name                : {name}")
    print(f"Hero Product Name   : {hero_product_name}")
    print(f"Category            : {category}")
    print(f"Catalog Price       : ₹{price}")
    print(f"Product Price       : ₹{product_price}")

    print(f"\nDescription:")
    print(description)

    print(f"\nRating Information:")
    print(f"Average Rating      : {rating}")
    print(f"Review Count        : {review_count}")
    print(f"Rating Count        : {rating_count}")

    print(f"\nProduct Details:")
    print(full_details)

    print(f"\nImage URL:")
    print(image_url)

    print("\n")
```
